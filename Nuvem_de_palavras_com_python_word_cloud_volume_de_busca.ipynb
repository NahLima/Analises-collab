```
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
```

```
#criar dataframe
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ-F41t38pSdbQl-7kB8621mvKv0O8pT-ArffgY8wvgLl22sbUI1E-xpZwGUS1Z6Vse9CFlpsyJVBn5/pub?gid=0&single=true&output=csv')
```

```
data = dict(zip(df['Keyword'].tolist(), df['Searches: Apr 2020'].tolist()))

print(data)
```

Nuvem de palavras

```
wc = WordCloud
wc.generate_from_frequencies(data)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```
wc = WordCloud(width=800, height=400, max_words=200).generate_from_frequencies(data)
```

Parametros da WordCloud

* background_color = Color of background
* mac_words = The maximum number of unique words used
* stopwords = stopword list
* max_font_size = Maximum font size
* random_state = To ensure that random numbers are generated in the
same order, so the results will be the same even if generated several times
* width = width size of the output
* height = height size of the output

```
wc = WordCloud(background_color="white", max_words=2000,
               max_font_size=256,
               random_state=42, width=1000, height=1000)
wc.generate_from_frequencies(data)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```
mask = np.array(Image.open('yamaha-c7x-sh2-silent-grand-piano-polished-ebony-p41146-80387_image.jpg')) #nome do arquivo da imagem
wc = WordCloud(font_path=font_path,
               mask=mask, background_color="white",
               max_words=2000, max_font_size=256,
               random_state=42, width=mask.shape[1],
               height=mask.shape[0])
wc.generate_from_frequencies(data)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```
from wordcloud import ImageColorGenerator
```

```
mask_colors = ImageColorGenerator(mask)
wc = WordCloud(font_path=font_path,
               mask=mask, background_color="white",
               max_words=2000, max_font_size=256,
               random_state=42, width=mask.shape[1],
               height=mask.shape[0], color_func=mask_colors)
wc.generate_from_frequencies(data)
plt.figure( figsize=(80,40) ) #ajustar tamanho imagem
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.savefig('sorveteice.jpg')
```
